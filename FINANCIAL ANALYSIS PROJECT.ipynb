```python
#IMPORTING THE NECESSARY LIBRARIES
import pandas as pd
import numpy as np
```

```python
#IMPORTING THE DATASET
finance_data =pd.read_csv('Finance_data.csv')
```

```python
#CREATING A DATAFRAME FROM THE DATASET
finance_data=pd.DataFrame(finance_data)
#CALLING OUT THE FIRST FIVE ROWS OF THE DATAFRAME
finance_data.head()
```

```python
#LENGTH OR THE NUMBER OF COLUMN OF THE DATASET
len(finance_data)
```

```python
#CHECKING OUT FOR MISSING VALUES
finance_data.isnull().sum()
```

```python
#CHECKING THE INFORMATION OF THE DATASET
finance_data.info()
```

```python
#IMPORTING THE NECESSARY LIBRARIES FOR DATA MANIPULATION AND VISUALIZATION
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
#INDICATING THE KIND OF GRID COLOR FOR THE VISUALISATION
sns.set_style('darkgrid')
```

```python
#DROPING THE COLUMNS THAT WONT BE USEFUL FOR THE VISUALIZATION AND MAKING IT REFLECT
finance_data.drop(['Mutual_Funds','Equity_Market','Debentures','Government_Bonds','PPF','Fixed_Deposits','Gold'],axis=1,inplace =True)
```

```python
#REPLACING OF SOME SIGNS WITH ANOTHER
finance_data.replace(r'^\s*$', np.nan, regex =True)
```

```python
#DATA VISUALIZATION SHOWING THE HIGHEST AND LOWEST FREQUENCY OF THE GENDER
custom_palette = ["#1f77b4", "#ff7f0e", "#2ca02c"]  # List of hex color codes
sns.countplot(x='gender', data=finance_data, linewidth=2, palette=custom_palette, edgecolor='black')
plt.show()
```

```python
#SHOWING THE FREQUENCY OF DIFFERENT AGE PRESENT IN THE DATASET
sns.countplot(x='age', data=finance_data, linewidth=2, palette=custom_palette, edgecolor='black')
plt.show()
```

```python
#INSIGHT OF THE INVESTMENT AVENUE WITH RESPECT TO THE GENDER
plt.figure(figsize=(18,6))
plt.subplot(1,2,1)
sns.countplot(x=finance_data['gender'], hue=finance_data['Investment_Avenues'],palette='summer',linewidth=3,edgecolor='white')
plt.title('INVESTMENT AVENUE')


#INSIGHT OF THE STOCK MARKET WITH RESPECT TO THE GENDER
plt.subplot(1,2,2)
sns.countplot(x=finance_data['gender'], hue=finance_data['Stock_Marktet'],palette='hot',linewidth=3,edgecolor='white')
plt.title('STOCK MARKET')
plt.show()
```

```python
#VISUALIZATION ON THE HIGHEST FACTOR AFFECTING THEIR INVESTMENT
sns.countplot(x=finance_data['Factor'], palette='coolwarm', linewidth=2, edgecolor='black')
plt.show()
```

```python
#VISUALIZATION EXPLANING THE PURPOSE OF INVESTING
plt.figure(figsize=(16,8))
sns.pointplot(x='Purpose', y='age', data=finance_data, linestyles= '--',capsize=.3,color='blue')
plt.show()
```

```python
#INSIGHT SHOWING THE HIGHEST AND LOWEST DURATION FOR EACH GENDER INVESTMENT
plt.figure(figsize=(18,6))
plt.subplot(1,2,1)
sns.countplot(hue=finance_data['gender'],x=finance_data['Duration'],palette='viridis',linewidth=2,edgecolor='black')
plt.title('DURATION')

#INSIGHT SHOWING THE HIGHEST AND LOWEST INVESTMONIROR WETHER MONTHLY OR WEEKLY OR DAILY FOR EACH GENDER INVESTMENT
plt.subplot(1,2,2)
sns.countplot(hue= finance_data['gender'],x=finance_data['Invest_Monitor'], palette='seismic',linewidth=2,edgecolor='black')
plt.title('INVESTMENT MONITORING')
plt.show()
```

```python
#INSIGHT SHOWING THE HIGHEST PERCENTAGE OF THEIR OBJECTIVE FOR THE INVESTMENT
# a pie chart
plt.subplot(1, 2, 2)  # Create a subplot for pie chart
factor_counts = finance_data['Objective'].value_counts()
labels = factor_counts.index
sizes = factor_counts.values
colors = sns.color_palette('coolwarm', len(labels))
plt.pie(sizes, labels=labels, colors=colors, autopct='%1.1f%%', startangle=140)
plt.title('Distribution of Objective')

plt.tight_layout()  # Adjust layout to prevent overlapping
plt.show()
```
